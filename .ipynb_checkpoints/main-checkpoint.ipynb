```python
import scipy.io as sio
import numpy as np

base = "/content/drive/MyDrive/Colab Notebooks/25-2 Reinforce/Project/project-data-main/standard_driving_cycles/mat-data/"

# 개별 사이클 로드
cltc     = sio.loadmat(base + "CLTC_P.mat")["speed_vector"].squeeze()
wvuinter = sio.loadmat(base + "Standard_WVUINTER.mat")["speed_vector"].squeeze()
wvucity  = sio.loadmat(base + "Standard_WVUCITY.mat")["speed_vector"].squeeze()
hwfet    = sio.loadmat(base + "Standard_HWFET.mat")["speed_vector"].squeeze()

# 1) 논문 학습용: MixTrain = CLTC_P + WVUINTER
mix_train = np.concatenate([cltc, wvuinter])
sio.savemat(base + "MixTrain.mat", {"speed_vector": mix_train})
print("✅ MixTrain.mat 저장 완료")

# 2) 논문 평가용: MixValid = WVUCITY + HWFET
mix_valid = np.concatenate([wvucity, hwfet])
sio.savemat(base + "MixValid.mat", {"speed_vector": mix_valid})
print("✅ MixValid.mat 저장 완료")
```

```python
import numpy as np
import scipy.io as scio
import matplotlib.pyplot as plt
import os

# episode data 폴더
base_dir = "/root/Reinforcement/Project/test8_SAC_CS_Beta/MixTrain_w100_LR1e-03_v1/episode_data"

max_temps = []  # 저장할 리스트

# 0~99 episode
for ep in range(100):
    path = f"{base_dir}/data_ep{ep}.mat"
    
    if not os.path.exists(path):
        print(f"파일 없음: {path}")
        max_temps.append(np.nan)
        continue

    data = scio.loadmat(path)

    # tep_c가 없을 경우 예외 처리
    if 'tep_c' not in data:
        print(f"tep_c 없음: {path}")
        max_temps.append(np.nan)
        continue

    Tep_c = np.array(data['tep_c']).flatten()
    max_temps.append(Tep_c.max())

# 그래프 시각화
plt.figure(figsize=(12,5))
plt.plot(range(100), max_temps, marker='o')
plt.title("Max Tep_c per Episode (0~99)")
plt.xlabel("Episode")
plt.ylabel("Max Battery Core Temp (°C)")
plt.ylim(50,80)
plt.grid(True)
plt.show()
```

```python
import scipy.io as scio
import numpy as np
import os

# 파일 경로 설정 (사용자님이 주신 경로)
base_dir = r"/root/Reinforcement/Project/test8_SAC_CS_Beta/MixTrain_w100_LR1e-03_v1"
reward_path = os.path.join(base_dir, "reward.mat")
eq_h2_path = os.path.join(base_dir, "eq_h2.mat")

print(f"Checking files in: {base_dir}")

if os.path.exists(reward_path) and os.path.exists(eq_h2_path):
    try:
        # 1. 파일 로드
        reward_data = scio.loadmat(reward_path)
        eq_h2_data = scio.loadmat(eq_h2_path)

        # 2. 데이터 추출 (flatten으로 1차원 배열로 변환)
        ep_r_list = reward_data['reward'].flatten()
        eq_h2_list = eq_h2_data['eq_h2'].flatten()

        # 3. 베스트 에피소드 찾기 (보상이 가장 높은 인덱스)
        best_ep_idx = np.argmax(ep_r_list)
        
        # 4. 해당 인덱스의 값들 가져오기
        best_ep_r = ep_r_list[best_ep_idx]
        best_eq_h2 = eq_h2_list[best_ep_idx]

        # 5. 결과 출력
        print("-" * 30)
        print(f"Best Episode: {best_ep_idx}")  # 0부터 시작하는 인덱스 (에피소드 번호)
        print(f"Max Reward: {best_ep_r:.4f}")
        print(f"Equivalent H2 (g/100km): {best_eq_h2:.4f}")
        print("-" * 30)

    except Exception as e:
        print(f"Error reading files: {e}")
else:
    print("\n[ERROR] 파일이 존재하지 않습니다!")
    print(f"Missing: {reward_path}")
    print("학습이 완료되지 않아 파일이 생성되지 않았을 수 있습니다.")
    print("이 경우, 'episode_data' 폴더를 직접 읽는 코드를 사용해야 합니다.")
```

# 평가 결과 비교 (8, 온도 보상 전 vs 9, 온도 보상 후)

```python
import scipy.io as scio
import numpy as np
import matplotlib.pyplot as plt
import os
import glob

# 경로 설정
path_no_penalty = "/root/Reinforcement/Project/eva_SAC_CS_Beta/root/Reinforcement/Project/test8_SAC_CS_Beta/MixTrain_w100_LR1e-03_v1_86/episode_data"
path_with_penalty = "/root/Reinforcement/Project/eva_SAC_CS_Beta/root/Reinforcement/Project/test9_SAC_CS_Beta/MixTrain_w100_LR1e-03_v1_73/episode_data"

def load_data(path):
    files = glob.glob(os.path.join(path, "data_ep*.mat"))
    files.sort() # 에피소드 순서대로 정렬
    
    max_temps = []
    rewards = []
    h2_consumptions = []
    
    for f in files:
        try:
            data = scio.loadmat(f)
            # Max Temperature
            max_temps.append(np.max(data['tep_c']))
            
            # Total Reward
            if 'EMS_reward' in data:
                rewards.append(np.sum(data['EMS_reward']))
            
            # Equivalent H2 Consumption (g/100km)
            travel_km = data['travel'][0][-1] / 1000
            total_eq_h2 = np.sum(data['h2_equal'])
            h2_100km = (total_eq_h2 / travel_km) * 100 if travel_km > 0 else 0
            h2_consumptions.append(h2_100km)
            
        except Exception as e:
            print(f"Error reading {f}: {e}")
            
    return max_temps, rewards, h2_consumptions

# 데이터 로드
t_no, r_no, h_no = load_data(path_no_penalty)
t_yes, r_yes, h_yes = load_data(path_with_penalty)

# 결과 출력
print("----- Comparison Results (Average over 5 episodes) -----")
print(f"Metrics                 | No Penalty (test8) | With Penalty (test9)")
print(f"-------------------------------------------------------------")
print(f"Max Core Temp (°C)      | {np.mean(t_no):.2f}               | {np.mean(t_yes):.2f}")
print(f"Total Reward            | {np.mean(r_no):.2f}             | {np.mean(r_yes):.2f}")
print(f"Eq H2 Consump (g/100km) | {np.mean(h_no):.2f}             | {np.mean(h_yes):.2f}")
print(f"-------------------------------------------------------------")

# 시각화
# 시각화
episodes = range(len(t_no))
plt.figure(figsize=(15, 5))

# Last 5 episodes average
avg_t_no = np.mean(t_no[-5:])
avg_t_yes = np.mean(t_yes[-5:])
avg_r_no = np.mean(r_no[-5:])
avg_r_yes = np.mean(r_yes[-5:])
avg_h_no = np.mean(h_no[-5:])
avg_h_yes = np.mean(h_yes[-5:])

# 1. Max Temperature Comparison
plt.subplot(1, 3, 1)
plt.plot(episodes, t_no, 'r-o', label='SOC Penalty')
plt.plot(episodes, t_yes, 'b-o', label='SOC+Temp Penalty')
plt.axhline(y=58, color='g', linestyle='--', label='Penalty Threshold (58°C)')
plt.axhline(y=60, color='k', linestyle='--', label='Fail Limit (60°C)')

# Add average lines
plt.axhline(y=avg_t_no, color='r', linestyle='--', alpha=0.5)
plt.axhline(y=avg_t_yes, color='b', linestyle='--', alpha=0.5)
plt.text(0, avg_t_no, f'{avg_t_no:.2f}', color='r', va='bottom')
plt.text(0, avg_t_yes, f'{avg_t_yes:.2f}', color='b', va='top')

plt.title('Max Battery Core Temperature')
plt.xlabel('Episode')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid(True)

# 2. Reward Comparison
plt.subplot(1, 3, 2)
plt.plot(episodes, r_no, 'r-o', label='SOC Penalty')
plt.plot(episodes, r_yes, 'b-o', label='SOC+Temp Penalty')

# Add average lines
plt.axhline(y=avg_r_no, color='r', linestyle='--', alpha=0.5)
plt.axhline(y=avg_r_yes, color='b', linestyle='--', alpha=0.5)
plt.text(0, avg_r_no, f'{avg_r_no:.2f}', color='r', va='bottom')
plt.text(0, avg_r_yes, f'{avg_r_yes:.2f}', color='b', va='top')

plt.title('Total Reward')
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.legend()
plt.grid(True)

# 3. H2 Consumption Comparison
plt.subplot(1, 3, 3)
plt.plot(episodes, h_no, 'r-o', label='SOC Penalty')
plt.plot(episodes, h_yes, 'b-o', label='SOC+Temp Penalty')

# Add average lines
plt.axhline(y=avg_h_no, color='r', linestyle='--', alpha=0.5)
plt.axhline(y=avg_h_yes, color='b', linestyle='--', alpha=0.5)
plt.text(0, avg_h_no, f'{avg_h_no:.2f}', color='r', va='bottom')
plt.text(0, avg_h_yes, f'{avg_h_yes:.2f}', color='b', va='top')

plt.title('Equivalent H2 Consumption')
plt.xlabel('Episode')
plt.ylabel('H2 (g/100km)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import scipy.io as scio
import numpy as np
import matplotlib.pyplot as plt
import os
import glob

# 경로 설정
path_no_penalty = "/root/Reinforcement/Project/eva_SAC_CS_Beta/root/Reinforcement/Project/test8_SAC_CS_Beta/MixTrain_w100_LR1e-03_v1_86/episode_data"
path_with_penalty = "/root/Reinforcement/Project/eva_SAC_CS_Beta/root/Reinforcement/Project/test9_SAC_CS_Beta/MixTrain_w100_LR1e-03_v1_73/episode_data"

def load_episode_data(path, episode_num=0):
    """특정 에피소드의 전체 시계열 데이터를 로드합니다."""
    file_path = os.path.join(path, f"data_ep{episode_num}.mat")
    try:
        data = scio.loadmat(file_path)
        return data
    except Exception as e:
        print(f"Error reading {file_path}: {e}")
        return None

# 비교할 에피소드 번호 (보통 0번 에피소드가 대표적)
target_ep = 0

data_no = load_episode_data(path_no_penalty, target_ep)
data_yes = load_episode_data(path_with_penalty, target_ep)

if data_no and data_yes:
    # 데이터 추출 (flatten으로 1차원 배열 변환)
    # 1. Temperature
    temp_no = data_no['tep_c'].flatten()
    temp_yes = data_yes['tep_c'].flatten()
    
    # 2. SOC
    soc_no = data_no['SOC'].flatten()
    soc_yes = data_yes['SOC'].flatten()
    
    # 3. Battery Power (kW) - P_batt is in W in the mat file
    pbatt_no = data_no['P_batt'].flatten() / 1000
    pbatt_yes = data_yes['P_batt'].flatten() / 1000
    
    # 4. Fuel Cell Power (kW) - P_FCS is usually normalized or stored differently, let's check keys
    # 'P_fc' might be the key, or we can infer from h2_fcs if needed. 
    # Based on evaluate.py: 'P_fc' is in episode_info.
    pfc_no = data_no['P_fc'].flatten()
    pfc_yes = data_yes['P_fc'].flatten()

    # 시간 축 (초 단위, 1 step = 1 sec 가정)
    time_steps = range(len(temp_no))

    # Calculate Average metrics of last 5 episodes
    def get_avg_metric(path, key, func=np.mean, scale=1.0):
        files = glob.glob(os.path.join(path, "data_ep*.mat"))
        files.sort()
        last_5_files = files[-5:]
        values = []
        for f in last_5_files:
            try:
                d = scio.loadmat(f)
                val = func(d[key].flatten())
                values.append(val)
            except:
                pass
        return np.mean(values) * scale if values else 0

    # 1. Temp (Max)
    avg_temp_no = get_avg_metric(path_no_penalty, 'tep_c', np.max)
    avg_temp_yes = get_avg_metric(path_with_penalty, 'tep_c', np.max)

    # 2. SOC (Mean)
    avg_soc_no = get_avg_metric(path_no_penalty, 'SOC', np.mean)
    avg_soc_yes = get_avg_metric(path_with_penalty, 'SOC', np.mean)

    # 3. P_batt (Mean, scale=1/1000 for kW)
    avg_pbatt_no = get_avg_metric(path_no_penalty, 'P_batt', np.mean, 0.001)
    avg_pbatt_yes = get_avg_metric(path_with_penalty, 'P_batt', np.mean, 0.001)

    # 4. P_fc (Mean)
    avg_pfc_no = get_avg_metric(path_no_penalty, 'P_fc', np.mean)
    avg_pfc_yes = get_avg_metric(path_with_penalty, 'P_fc', np.mean)

    # 그래프 그리기
    plt.figure(figsize=(18, 12))

    # 1. Battery Core Temperature
    plt.subplot(2, 2, 1)
    plt.plot(time_steps, temp_no, 'r-', label='SOC Penalty')
    plt.plot(time_steps, temp_yes, 'b-', label='SOC+Temp Penalty')
    plt.axhline(y=58, color='g', linestyle='--', label='Penalty Threshold (58°C)')
    plt.axhline(y=60, color='k', linestyle='--', label='Fail Limit (60°C)')
    
    # Add average lines
    plt.axhline(y=avg_temp_no, color='r', linestyle='--', linewidth=2.5, alpha=0.8)
    plt.axhline(y=avg_temp_yes, color='b', linestyle='--', linewidth=2.5, alpha=0.8)
    plt.text(time_steps[-1], avg_temp_no, f'SOC Penalty: {avg_temp_no:.2f}', color='r', fontsize=12, va='bottom', ha='right', bbox=dict(facecolor='white', alpha=0.9, edgecolor='none'))
    plt.text(time_steps[-1], avg_temp_yes, f'SOC+Temp Penalty: {avg_temp_yes:.2f}', color='b', fontsize=12, va='top', ha='right', bbox=dict(facecolor='white', alpha=0.9, edgecolor='none'))

    plt.title(f'Battery Core Temperature (Episode {target_ep})')
    plt.xlabel('Time (s)')
    plt.ylabel('Temperature (°C)')
    plt.legend()
    plt.grid(True)

    # 2. SOC Trajectory
    plt.subplot(2, 2, 2)
    plt.plot(time_steps, soc_no, 'r-', label='SOC Penalty')
    plt.plot(time_steps, soc_yes, 'b-', label='SOC+Temp Penalty')
    
    # Add average lines
    plt.axhline(y=avg_soc_no, color='r', linestyle='--', linewidth=2.5, alpha=0.8)
    plt.axhline(y=avg_soc_yes, color='b', linestyle='--', linewidth=2.5, alpha=0.8)
    plt.text(time_steps[-1], avg_soc_no, f'SOC Penalty: {avg_soc_no:.4f}', color='r', fontsize=12, va='bottom', ha='right', bbox=dict(facecolor='white', alpha=0.9, edgecolor='none'))
    plt.text(time_steps[-1], avg_soc_yes, f'SOC+Temp Penalty: {avg_soc_yes:.4f}', color='b', fontsize=12, va='top', ha='right', bbox=dict(facecolor='white', alpha=0.9, edgecolor='none'))

    plt.title(f'SOC Trajectory (Episode {target_ep})')
    plt.xlabel('Time (s)')
    plt.ylabel('SOC')
    plt.legend()
    plt.grid(True)

    # 3. Battery Power Output
    plt.subplot(2, 2, 3)
    plt.plot(time_steps, pbatt_no, 'r-', label='SOC Penalty')
    plt.plot(time_steps, pbatt_yes, 'b-', label='SOC+Temp Penalty')
    
    # Add average lines
    plt.axhline(y=avg_pbatt_no, color='r', linestyle='--', linewidth=2.5, alpha=0.8)
    plt.axhline(y=avg_pbatt_yes, color='b', linestyle='--', linewidth=2.5, alpha=0.8)
    plt.text(time_steps[-1], avg_pbatt_no, f'SOC Penalty: {avg_pbatt_no:.2f}', color='r', fontsize=12, va='bottom', ha='right', bbox=dict(facecolor='white', alpha=0.9, edgecolor='none'))
    plt.text(time_steps[-1], avg_pbatt_yes, f'SOC+Temp Penalty: {avg_pbatt_yes:.2f}', color='b', fontsize=12, va='top', ha='right', bbox=dict(facecolor='white', alpha=0.9, edgecolor='none'))

    plt.title(f'Battery Power Output (Episode {target_ep})')
    plt.xlabel('Time (s)')
    plt.ylabel('Power (kW)')
    plt.legend()
    plt.grid(True)

    # 4. Fuel Cell Power Output
    plt.subplot(2, 2, 4)
    plt.plot(time_steps, pfc_no, 'r-', label='SOC Penalty')
    plt.plot(time_steps, pfc_yes, 'b-', label='SOC+Temp Penalty')
    
    # Add average lines
    plt.axhline(y=avg_pfc_no, color='r', linestyle='--', linewidth=2.5, alpha=0.8)
    plt.axhline(y=avg_pfc_yes, color='b', linestyle='--', linewidth=2.5, alpha=0.8)
    plt.text(time_steps[-1], avg_pfc_no, f'SOC Penalty: {avg_pfc_no:.2f}', color='r', fontsize=12, va='bottom', ha='right', bbox=dict(facecolor='white', alpha=0.9, edgecolor='none'))
    plt.text(time_steps[-1], avg_pfc_yes, f'SOC+Temp Penalty: {avg_pfc_yes:.2f}', color='b', fontsize=12, va='top', ha='right', bbox=dict(facecolor='white', alpha=0.9, edgecolor='none'))

    plt.title(f'Fuel Cell Power Output (Episode {target_ep})')
    plt.xlabel('Time (s)')
    plt.ylabel('Power (kW)')
    plt.ylim(-10,70)

    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    print("Graphs generated successfully.")
    
    # 추가 분석: 온도가 58도를 넘는 구간 비율
    over_58_no = np.sum(temp_no > 58) / len(temp_no) * 100
    over_58_yes = np.sum(temp_yes > 58) / len(temp_yes) * 100
    
    print(f"\n[Analysis]")
    print(f"Time spent over 58°C:")
    print(f"  - SOC Penalty: {over_58_no:.2f}%")
    print(f"  - SOC+Temp Penalty: {over_58_yes:.2f}%")

else:
    print("Failed to load data. Please check the paths.")

```

```python
def collect_all_temps(path):
    temps = []
    files = glob.glob(os.path.join(path, "data_ep*.mat"))
    files.sort()
    for f in files:
        data = scio.loadmat(f)
        temps.append(data['tep_c'].flatten())
    return np.concatenate(temps)

all_temp_no = collect_all_temps(path_no_penalty)
all_temp_yes = collect_all_temps(path_with_penalty)

plt.figure(figsize=(6,4))
plt.hist(all_temp_no, bins=40, alpha=0.5, label='No Penalty', density=True)
plt.hist(all_temp_yes, bins=40, alpha=0.5, label='With Penalty', density=True)
plt.axvline(58, color='g', linestyle='--', label='58°C')
plt.axvline(60, color='k', linestyle='--', label='60°C')
plt.xlabel('Battery Core Temperature (°C)')
plt.ylabel('Density')
plt.title('Temperature Distribution')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()


```

```python
def load_vec(path, name):
    d = scio.loadmat(os.path.join(path, f"{name}.mat"))
    return np.array(d[name]).squeeze()

# 1) 에피소드별 curve들
reward8 = load_vec(path8, "reward")
reward9 = load_vec(path9, "reward")
money8  = load_vec(path8, "money")
money9  = load_vec(path9, "money")
eqh2_8  = load_vec(path8, "eq_h2_100")
eqh2_9  = load_vec(path9, "eq_h2_100")

episodes = np.arange(1, len(reward8)+1)

plt.figure(figsize=(15,4))

plt.subplot(1,3,1)
plt.plot(episodes, reward8, label='No Temp Penalty')
plt.plot(episodes, reward9, label='With Temp Penalty')
plt.xlabel('Episode'); plt.ylabel('Reward'); plt.title('Episode Avg Reward')
plt.grid(True, alpha=0.3); plt.legend()

plt.subplot(1,3,2)
plt.plot(episodes, money8, label='No Temp Penalty')
plt.plot(episodes, money9, label='With Temp Penalty')
plt.xlabel('Episode'); plt.ylabel('Money (¥/100km)')
plt.title('Money per 100 km')
plt.grid(True, alpha=0.3); plt.legend()

plt.subplot(1,3,3)
plt.plot(episodes, eqh2_8, label='No Temp Penalty')
plt.plot(episodes, eqh2_9, label='With Temp Penalty')
plt.xlabel('Episode'); plt.ylabel('Eq H2 (g/100km)')
plt.title('Eq H2 per 100 km')
plt.grid(True, alpha=0.3); plt.legend()

plt.tight_layout()
plt.show()
```
